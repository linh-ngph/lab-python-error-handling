   - If the user enters an invalid product name (e.g., a product name that is not in the inventory), or that doesn't have stock available, display an error message and ask them to re-enter the product name. *Hint: you will need to pass inventory as a parameter*
   - Use a try-except block to handle the error and continue prompting the user until a valid product name is entered.

4. Test your code by running the program and deliberately entering invalid quantities and product names. Make sure the error handling mechanism works as expected.


````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {}
customer_orders = set()

def initialize_inventory(products):
    global inventory
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
                    continue
                else:
                    inventory[product] = quantity
                    valid_input = True 
            except:
                 print("Invalid input. Please enter a valid quantity.")
    return inventory
                    

def get_customer_orders():
    global customer_orders
    valid_number_order = False
    while not valid_number_order:
        try:
            number_customer_orders = int(input("How many products would you like to order?: "))
            if number_customer_orders < 0:
                print("Quantity cannot be negative. Please enter a valid quantity.")
                continue
            else:
                for _ in range(number_customer_orders):
                    valid_name_product = False  # Reset the flag inside the loop
                    while not valid_name_product:
                        try:
                            to_order = input(f"Our products are: {products} \n Enter the names of the products you want to order: ")
                            if to_order not in products:
                                print("A product name that is not in the inventory. Try again.")
                                continue
                            else:
                                item_to_order = [item.strip().lower() for item in to_order.split(',')]
                                for item in item_to_order:
                                    if item in products: 
                                        customer_orders.add(item)
                                    else:
                                        print(f"These products: {item} is not in inventory. Try again.")
                                        continue
                                valid_name_product = True  # Set to True when input is valid
                        except:
                            print("Invalid input. Please enter a valid product name.")  # This block was misaligned
                valid_number_order = True  # Once all products are entered, break out of the main loop
        except:
            print("Invalid input. Please enter a valid number.")
    return customer_orders


def price(orders):
    total_price = 0
    for item in orders:
        valid_price = False
        while not valid_price:
            try:
                price_to_ask = int(input(f"Enter the price of {item}: "))
                if price_to_ask < 0:
                    print("Price cannot be negative. Please enter a valid price.")
                else:
                    total_price += price_to_ask
                    valid_price = True
            except ValueError:
                print("Invalid input. Please enter a valid price.")
            
    return total_price


def update_inventory(customer_orders, inventory):
    updated_inventory = {
        item: inventory.get(item, 0) - 1
        for item in customer_orders
        if inventory.get(item, 0) > 0
    }
    inventory.update({item: quantity for item, quantity in updated_inventory.items() if quantity > 0})

def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    total_products = len(products)
    percentage_unique = (total_products_ordered / total_products) * 100 if total_products > 0 else 0
    return total_products_ordered, percentage_unique

def print_order_statistics(order_statistics):
    total_products_ordered, percentage_unique = order_statistics
    print(f"Total products ordered: {total_products_ordered}\nPercentage unique products ordered: {percentage_unique:.2f}%")

def print_updated_inventory(inventory):
    print("Updated Inventory:")
    for item, quantity in inventory.items():
        print(f"{item}: {quantity}")

def print_total_price(total_price):
    print(f"Total price: {total_price}$")


initialize_inventory(products)
customer_orders = get_customer_orders()

update_inventory(customer_orders, inventory)
order_statistics = calculate_order_statistics(customer_orders, products)
price_customer_orders = price(customer_orders)

print_order_statistics(order_statistics)
print_updated_inventory(inventory)
print_total_price(price_customer_orders)
````
